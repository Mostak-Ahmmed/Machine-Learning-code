```python
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

xs = np.array([[0], [1], [2], [3]])
ys = np.array([0, 1, 1, 0])

# Split the data into training and test sets (80/20 split)
xs_train, xs_test, ys_train, ys_test = train_test_split(xs, ys, test_size=0.2, random_state=42)
```

```python

hidden_layer_configurations = [
    (3), (5), (10),
    (3, 3), (5, 5), (10, 10)
]

for layers in hidden_layer_configurations:
    model = MLPClassifier(
        activation='logistic',        # Logistic activation function
        max_iter=10000,               # Same number of iterations
        learning_rate_init=0.0001,    # Same learning rate
        hidden_layer_sizes=layers     # Different layer configurations
    )
    model.fit(xs_train, ys_train)
    predictions = model.predict(xs_test)
    accuracy = accuracy_score(ys_test, predictions)
    print(f'Hidden Layer Configuration: {layers} | Accuracy: {accuracy:.4f}')




# iterations = [100, 1000, 5000, 10000]
# for max_iter in iterations:
#     model = MLPClassifier(
#         activation='logistic',        # Logistic activation function
#         max_iter=max_iter,           # Varying the number of iterations
#         learning_rate_init=0.0001,    # Use same learning rate
#         hidden_layer_sizes=(3)        # One hidden layer with 3 neurons
#     )
#     model.fit(xs_train, ys_train)
#     predictions = model.predict(xs_test)
#     accuracy = accuracy_score(ys_test, predictions)
#     print(f'Max Iterations: {max_iter} | Accuracy: {accuracy:.4f}')



# learning_rates = [0.1, 0.01, 0.001, 0.0001]
# for lr in learning_rates:
#     model = MLPClassifier(
#         activation='logistic',       # Logistic activation function
#         max_iter=10000,              # Use the same iteration count
#         learning_rate_init=lr,       # Varying the learning rate
#         hidden_layer_sizes=(3)       # One hidden layer with 3 neurons
#     )
#     model.fit(xs_train, ys_train)
#     predictions = model.predict(xs_test)
#     accuracy = accuracy_score(ys_test, predictions)
#     print(f'Learning Rate: {lr} | Accuracy: {accuracy:.4f}')




# activation_functions = ['identity', 'logistic', 'tanh', 'relu']
# for activation in activation_functions:
#     model = MLPClassifier(
#         activation=activation,       # Change activation function
#         max_iter=10000,              # Use same iteration count for now
#         learning_rate_init=0.0001,   # Use same learning rate
#         hidden_layer_sizes=(3)       # Use one hidden layer with 3 neurons
#     )
#     model.fit(xs_train, ys_train)
#     predictions = model.predict(xs_test)
#     accuracy = accuracy_score(ys_test, predictions)
#     print(f'Activation: {activation} | Accuracy: {accuracy:.4f}')




# from sklearn.neural_network import MLPClassifier

# # Define the model with the parameters you mentioned
# model = MLPClassifier(
#     activation='relu',        # Use logistic activation function (sigmoid)
#     max_iter=10000,               # Maximum number of iterations for training
#     learning_rate_init=0.001,    # Initial learning rate
#     hidden_layer_sizes=(5)        # One hidden layer with 3 units
# )

# # Fit the model with the training data (xs) and labels (ys)
# model.fit(xs, ys)
```

```python
import matplotlib.pyplot as plt

# Example: Activation function results
activation_results = {
    'identity': 0.5000,
    'logistic': 0.5000,
    'tanh': 0.5000,
    'relu': 1.0000
}

# Bar plot for activation function comparison
plt.figure(figsize=(8, 5))
plt.bar(activation_results.keys(), activation_results.values(), color=['red', 'blue', 'green', 'orange'])
plt.xlabel('Activation Function')
plt.ylabel('Accuracy')
plt.title('Comparison of Activation Functions')
plt.show()
```

```python
# Example: Learning rate results
learning_rate_results = {
    0.1: 0.5000,
    0.01: 1.0000,
    0.001: 1.0000,
    0.0001: 1.0000
}

# Bar plot for learning rate comparison
plt.figure(figsize=(8, 5))
plt.bar([str(lr) for lr in learning_rate_results.keys()], learning_rate_results.values(), color=['lightblue', 'blue', 'green', 'purple'])
plt.xlabel('Learning Rate')
plt.ylabel('Accuracy')
plt.title('Comparison of Learning Rates')
plt.show()
```

```python
#Iteration results
iterations_results = {
    100: 0.5000,
    1000: 1.0000,
    5000: 1.0000,
    10000: 1.0000
}

# Bar plot for iterations comparison
plt.figure(figsize=(8, 5))
plt.bar([str(iteration) for iteration in iterations_results.keys()], iterations_results.values(), color=['lightgreen', 'green', 'darkgreen', 'yellowgreen'])
plt.xlabel('Max Iterations')
plt.ylabel('Accuracy')
plt.title('Comparison of Max Iterations')
plt.show()
```

```python
# Example: Layer configuration results
layer_results = {
    (3): 1.0000,
    (5): 1.0000,
    (10): 1.0000,
    (3, 3): 1.0000,
    (5, 5): 1.0000,
    (10, 10): 1.0000
}

# Bar plot for layer configurations comparison
plt.figure(figsize=(10, 6))
plt.bar([str(layers) for layers in layer_results.keys()], layer_results.values(), color=['lightblue', 'blue', 'green', 'orange', 'purple', 'red'])
plt.xlabel('Hidden Layer Configuration')
plt.ylabel('Accuracy')
plt.title('Comparison of Hidden Layer Configurations')
plt.xticks(rotation=45)
plt.show()
```

```python
# Example: Learning rate vs accuracy (Line Plot)
learning_rate_results = {
    0.1: 0.5000,
    0.01: 1.0000,
    0.001: 1.0000,
    0.0001: 1.0000
}

# Line plot for learning rate vs accuracy
plt.figure(figsize=(8, 5))
plt.plot(list(learning_rate_results.keys()), list(learning_rate_results.values()), marker='o', color='blue', linestyle='-', linewidth=2)
plt.xlabel('Learning Rate')
plt.ylabel('Accuracy')
plt.title('Learning Rate vs Accuracy')
plt.grid(True)
plt.show()
```
