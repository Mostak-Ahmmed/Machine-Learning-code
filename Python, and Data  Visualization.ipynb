Write a Python program that calculates the frequency of each character in a given
input string.

```python
import matplotlib.pyplot as plt

# Function to count the frequency of each character
def count_character_frequency(input_string):
    # Create an empty dictionary to store character counts
    char_frequency = {}

    # Iterate through each character in the string
    for char in input_string:
        # If the character is already in the dictionary, increment its count
        if char in char_frequency:
            char_frequency[char] += 1
        # If the character is not in the dictionary, add it with a count of 1
        else:
            char_frequency[char] = 1

    # Return the character frequency dictionary
    return char_frequency

# Input string
input_string = input("Enter a string: ")

# Get the character frequencies
char_frequency = count_character_frequency(input_string)

# Print the frequencies
for char, count in char_frequency.items():
    print(f"'{char}': {count}")

# Visualization using a bar chart
chars = list(char_frequency.keys())  # List of characters
counts = list(char_frequency.values())  # List of character counts

# Create a bar chart
plt.bar(chars, counts)

# Labeling the chart
plt.xlabel('Characters')
plt.ylabel('Frequency')
plt.title('Character Frequency Distribution')

# Display the chart
plt.show()
```

You are given the task of calculating the total tax amount for a person’s income
based on the following tax brackets: - Income up to $10,000 is taxed at a rate of
5%. - Income from $10,001 to $50,000 is taxed at a rate of 10% - Income from
$50,001 to $100,000 is taxed at a rate of 20%. - Income over $100,000 is taxed at
a rate of 30%.

```python
import matplotlib.pyplot as plt

# Function to calculate the tax based on income
def calculate_tax(income):
    tax = 0
    tax_brackets = []

    if income <= 10000:
        tax = income * 0.05
        tax_brackets = [(0, income, 0.05, income * 0.05)]  # (start, end, rate, tax amount)
    elif income <= 50000:
        tax = 10000 * 0.05 + (income - 10000) * 0.10
        tax_brackets = [
            (0, 10000, 0.05, 10000 * 0.05),
            (10000, income, 0.10, (income - 10000) * 0.10)
        ]
    elif income <= 100000:
        tax = 10000 * 0.05 + 40000 * 0.10 + (income - 50000) * 0.20
        tax_brackets = [
            (0, 10000, 0.05, 10000 * 0.05),
            (10000, 50000, 0.10, 40000 * 0.10),
            (50000, income, 0.20, (income - 50000) * 0.20)
        ]
    else:
        tax = 10000 * 0.05 + 40000 * 0.10 + 50000 * 0.20 + (income - 100000) * 0.30
        tax_brackets = [
            (0, 10000, 0.05, 10000 * 0.05),
            (10000, 50000, 0.10, 40000 * 0.10),
            (50000, 100000, 0.20, 50000 * 0.20),
            (100000, income, 0.30, (income - 100000) * 0.30)
        ]

    return tax, tax_brackets

# Input income
income = float(input("Enter your income: $"))

# Calculate the tax and tax brackets
tax, tax_brackets = calculate_tax(income)

# Print the total tax owed
print(f"Total tax owed: ${tax:.2f}")

# Visualization of tax breakdown
bracket_labels = [f"${start} - ${end}" for start, end, _, _ in tax_brackets]
tax_amounts = [tax_amount for _, _, _, tax_amount in tax_brackets]

# Create a bar chart for tax amounts
plt.bar(bracket_labels, tax_amounts, color='green')

# Labeling the chart
plt.xlabel('Income Bracket')
plt.ylabel('Tax Amount ($)')
plt.title('Tax Distribution Based on Income Brackets')

# Display the chart
plt.show()
```
